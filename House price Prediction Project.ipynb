```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
labels = data['Price']
conv_dates = [1 if values == 2014 else 0 for values in data.Date ]
data['Date'] = conv_dates
train1 = data.drop(['ID', 'Price'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.10,random_state =2)
```

```python
reg.fit(x_train,y_train)
```

```python
#Acurracy after only using Linear Regression
reg.score(x_test,y_test)
```

```python
from sklearn import ensemble
```

```python
params = {'n_estimators' : 400, 'max_depth' : 5, 'min_samples_split' : 2,
          'learning_rate' : 0.1, 'loss' : 'ls'}
grb = ensemble.GradientBoostingRegressor(**params)
```

```python
grb.fit(x_train, y_train)
```

```python
#Accuracy after Gradient Booster
grb.score(x_test,y_test)
```

```python
#using slicing
#predicts the prices of the houses with the accuracy of 91%
y_pred = reg.predict(x_test[0:4])
y_pred
```

```python
from sklearn import metrics
```

```python
predictions = grb.predict(x_test) 
```

```python
plt.figure(figsize=(6,8))
plt.scatter(y_test,predictions)
```

```python
sns.displot((y_test-predictions),bins=50); 
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions)) #Mean absolute error
print('MSE:', metrics.mean_squared_error(y_test, predictions)) #mean squared error
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions))) #root mean squared error
```
